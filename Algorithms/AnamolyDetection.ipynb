```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
df = pd.read_csv('tips.csv')
print(df.head())
```

```python
# Convert categorical columns to numeric (if necessary)
# Here, assuming 'sex' is a categorical column, we can use one-hot encoding
df_encoded = pd.get_dummies(df, drop_first=True)

# Plotting a scatter plot of the first two numeric columns
plt.scatter(df_encoded.iloc[:, 0], df_encoded.iloc[:, 1])
plt.xlabel(df_encoded.columns[0])
plt.ylabel(df_encoded.columns[1])
plt.show()
```

```python
# Fit the IsolationForest model
clf = IsolationForest(contamination='auto')
clf.fit(df_encoded)

# Predict anomalies
predictions = clf.predict(df_encoded)

# Output the predictions
print(predictions)
```

```python
import numpy as np
index=np.where(predictions<0)
index
```

```python
x=df.values
```

```python
index=np.where(predictions<0)
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors="r")
```
